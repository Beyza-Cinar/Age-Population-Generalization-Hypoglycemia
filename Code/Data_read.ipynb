```python
# necessary imports
import pandas as pd
import numpy as np
from datetime import timedelta
import polars as pl
 
from data_integration import combine_data, read_data 
```

## Read Data  

```python
"""
    Calls the "read_data()" function from file "data_integration.py"
    Calls the "combine_data()" function from file "data_integration.py"
    Parameter:
    - 1: combines only glucose data
    - 2: combines glucose and dempgraphics
    - 3: combines glucose and heartrate
    Output: This function returns the intrgrated dataset included the restricted databases if downloaded and added to folder "datasets for T1D"
 """
# reads all datasets, if they are stored in the required folder Data/datasets_for_T1D...
# DiaData includes two restricted datasets, these should be downloaded from the publishers and shored in the folder
restricted_list_5 = read_data(read_all = False, frequency = "15min")
# integrates the restricted datasets with modus = 2
df_5min = combine_data(2, restricted_list_5, columns_to_check=["Age", "Sex"])

# combine with not restricted data
# This dataset can be downloaded from https://doi.org/10.5281/zenodo.16874128 and should be stored in Data/datasets_for_T1D 
sdb_I_5min = pl.read_csv("Data/datasets_for_T1D/SDBI_5min.csv")
sdb_I_5min  = sdb_I_5min.sort(["PtID", "ts"])
sdb_I_5min = sdb_I_5min.to_pandas()
# converts the timestamp column to datetime
sdb_I_5min["ts"] = pd.to_datetime(sdb_I_5min["ts"])

# combines both dataframes to have the final integrated dataset 
df_5min = pd.concat([df_5min, sdb_I_5min])
```

```python
# reads all datasets, if they are stored in the required folder Data/datasets_for_T1D...
restricted_list_15 = read_data(read_all = False, frequency = "15min")
# integrates the restricted datasets with modus = 2
df_15min = combine_data(2, restricted_list_15, columns_to_check=["Age", "Sex"])

# combine with not restricted data
# This dataset can be downloaded from https://doi.org/10.5281/zenodo.16874128 and should be stored in Data/datasets_for_T1D
sdb_I_15min = pl.read_csv("Data/datasets_for_T1D/SDBI_15min.csv")
sdb_I_15min  = sdb_I_15min.sort(["PtID", "ts"])
sdb_I_15min = sdb_I_15min.to_pandas()
# converts the timestamp column to datetime
sdb_I_15min["ts"] = pd.to_datetime(sdb_I_15min["ts"])

# combines both dataframes to have the final integrated dataset 
df_15min = pd.concat([df_15min, sdb_I_15min])
```

```python
# converts to polars for faster writing
schema = {
    "Age": pl.Utf8,
}
df_5min["Age"] = df_5min["Age"].astype(str)
df_polars_5 = pl.from_pandas(df_5min, schema_overrides=schema)
```

```python
# converts to polars for faster writing
schema = {
    "Age": pl.Utf8,
}
df_15min["Age"] = df_15min["Age"].astype(str)
df_polars_15 = pl.from_pandas(df_15min, schema_overrides=schema)
```

```python
# saves both dataframes
df_polars_5.write_csv("Data/SDBII_5.csv")
df_polars_15.write_csv("Data/SDBII_15.csv")
```
