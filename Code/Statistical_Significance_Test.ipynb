```python
# First we used pairwise wilcoxon comparisons for each age group separately

groups = ["Children","Teenagers","Adults","Seniors"]

for g in range(4):
    print("\n", groups[g])
    print("Corrected alpha:", alpha)

    stat,p = wilcoxon(model1[g], model2[g])
    print("Model1 vs Model2:", p, "Significant:", p < alpha)

    stat,p = wilcoxon(model3[g], model4[g])
    print("Model3 vs Model4:", p, "Significant:", p < alpha)

    stat,p = wilcoxon(model1[g], model3[g])
    print("Model1 vs Model3:", p, "Significant:", p < alpha)

    stat,p = wilcoxon(model2[g], model4[g])
    print("Model2 vs Model4:", p, "Significant:", p < alpha)
```

```python
# Second, the results per model were concatenated and population based performance was compared
m1 = np.concatenate(model1)
m2 = np.concatenate(model2)
m3 = np.concatenate(model3)
m4 = np.concatenate(model4)

tests = {
    "Model1 vs Model2": wilcoxon(m1, m2).pvalue,
    "Model3 vs Model4": wilcoxon(m3, m4).pvalue,
    "Model1 vs Model3": wilcoxon(m1, m3).pvalue,
    "Model2 vs Model4": wilcoxon(m2, m4).pvalue
}

for name, p in tests.items():
    significant = p < alpha
    print(name, "p =", p, "significant:", significant)
```
